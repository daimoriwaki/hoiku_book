```python
# make region_json to DataFrame

def make_region_df(regions_json):
    region_df = pd.DataFrame(columns=['region', 'prefecture', 'city'])
    for region, region_dict in regions_json.items():
        for prefecture, cities in region_dict.items():
            for city in cities:
                region_df = region_df.append(
                    pd.Series([region, prefecture, city], index=region_df.columns), ignore_index=True)
    return region_df
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Load shapefile
shapefile_path = "/Users/a14880/Downloads/N03-20240101_GML/N03-20240101.shp"
gdf = gpd.read_file(shapefile_path)



# Merge datasets
gdf = gdf.merge(region_df, left_on=["N03_001", "N03_004"], right_on=["prefecture", "city"], how="left")

gdf['region'].fillna(0, inplace = True)

gdf['region'] = gdf['region'].astype(int)

# Plot map with 地域区分
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
gdf.plot(column='region', cmap='Greys', legend=True, ax=ax)
ax.set_title("地域区分 by City Level in Japan")
plt.show()
```

```python
from matplotlib.colors import Normalize, LinearSegmentedColormap

greys = plt.cm.Greys
custom_cmap = LinearSegmentedColormap.from_list("shifted_greys", greys(range(50, 256)))  # Start from grey

# Normalize to scale values from 0 to 20
norm = Normalize(vmin=0, vmax=20)

# Plot map with adjusted greyscale
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
gdf.plot(column='region', cmap=custom_cmap, norm=norm, ax=ax)  # Removed legend=True
ax.set_title("地域区分 by City Level in Japan")
plt.show()
```

```python
from matplotlib.patches import Patch

# Define styles for each region value
hatch_styles = {
    20: {'facecolor': 'black', 'hatch': None},   # Black
    16: {'facecolor': 'grey', 'hatch': '..'},    # Pointed
    15: {'facecolor': 'grey', 'hatch': '..'},    # Pointed
    12: {'facecolor': 'grey', 'hatch': '..'},    # Pointed
    10: {'facecolor': 'grey', 'hatch': '..'},    # Pointed
    6: {'facecolor': 'grey', 'hatch': '///'},   # Slanted lines
    3: {'facecolor': 'grey', 'hatch': '///'},   # Slanted lines
    0: {'facecolor': 'grey', 'hatch': None}     # Grey
}

# Plotting
fig, ax = plt.subplots(1, 1, figsize=(15, 10))

# Iterate through unique regions and plot with specific styles
for region_value, style in hatch_styles.items():
    subset = gdf[gdf['region'] == region_value]
    subset.plot(
        ax=ax,
        color=style['facecolor'],
        edgecolor='black',
        hatch=style['hatch'],
        linewidth=0.5
    )a

# Add a custom legend
legend_elements = [
    Patch(facecolor='black', label='20 (Black)'),
    Patch(facecolor='grey', hatch='..', label='16, 15, 12, 10 (Pointed)'),
    Patch(facecolor='grey', hatch='///', label='6, 3 (slanted)'),
    Patch(facecolor='grey', label='0 (Grey)')
]
ax.legend(handles=legend_elements, loc='upper left', title='Region Styles')

ax.set_title("地域区分 by City Level in Japan")
plt.show()
```
